```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("iris.data", delimiter=",")
```

```python
df.head()
```

```python
df.info()
df["Rotulo"] = df["Rotulo"].astype(str)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,10))
ax = sns.boxplot(x = "Rotulo", y = "comPetala", data = df)
```

```python
plt.figure(figsize=(10,10))
sns.pairplot(df, hue="Rotulo")
```

```python
x = df.values[:,:-1]#Conjunto de atributos
y = df.values[:,-1]#Rotulo do meu conjunto de dados

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.3)
```

```python
#### EXECUÇÃO DO MODELO DE REGRESSÂO LOGISTICA####
modelo = LogisticRegression()
modelo.fit(xtrain,ytrain)
```

```python
y_Predicao = modelo.predict(xtest)

#y_Predicao
tmp = np.vstack((y_Predicao, ytest)).T
pd.DataFrame(tmp, columns=["valor Predito","valor Real"])
```
